```python
# Getting Ready to use Dataset
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0, test_size=0.3)
```

```python
print('x_train size: \n', x_train.shape)
print('x_test size: \n', x_test.shape)
print('y_train size: \n', y_train.shape)
print('y_test size: \n', y_test.shape)
```

```python
# Data Visualization
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# DataFrame(data, columnnames)
dataFrame= pd.DataFrame(x_train, columns=iris_dataset['feature_names'])
```

```python
pd.plotting.scatter_matrix(dataFrame, 
                           figsize=(10,10),
                           marker='o',
                           c=y_train,
                           alpha=0.8
                           )
```

```python
# first Machine Learning Model: K-Neighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

```python
# New Predict Sample Data
x_new = np.array([[5,2.9,1,0.2]])
prediction = knn.predict(x_new)
```

```python
print('Prediction Value: ', prediction)
print('Predict Target Name:',iris_dataset['target_names'][prediction])
```

```python
# Model Evaluation
y_predict = knn.predict(x_test)
```

```python
print('Result of the Test Set:\n',y_predict)
```

```python
print('Test Result Percentage: ',round(np.mean(y_predict == y_test) *100,2),'%')
```
