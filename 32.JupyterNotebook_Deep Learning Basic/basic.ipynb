```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn-whitegrid')
```

```python
def linear_function(x):
    a = 0.5
    b= 2
    return b + np.multiply(x,a)
```

```python
linear_function(5)
```

```python
a = np.array([3,4,5,6,7,8])
linear_function(a)
```

```python
b= np.arange(-5,5,0.1)
```

```python
linear_function(b)
```

```python
x = np.arange(-5,5,0.1)
y=linear_function(x)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

```python
def my_func(x):
    a = 1
    b = -3
    c = 10
    return a*x**2+b*x+c
```

```python
x= np.arange(-10,10,0.1)
y=my_func(x)
y_min = 100
x_min = 0
for i in range(len(x)):
    y_current = my_func(x[i])
    if y_current<y_min:
        y_min = y_current
        x_min= x[i]
     

plt.scatter(x_min,y_min, c='red')

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('My Function')
```
